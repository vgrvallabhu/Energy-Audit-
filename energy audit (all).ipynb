```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
ds =pd.read_csv('audit.csv')
ds
df=ds
```

```python
x = ds.iloc[:,1:-1].values
y = ds.iloc[:,-1:].values
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,0] = le.fit_transform(x[:,0])
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Naive Bayes classifier

```python
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB() 
```

```python
gnb.fit(x_train,y_train) 
```

```python
y_pred = gnb.predict(x_test) 
```

```python
from sklearn import metrics 
g= metrics.accuracy_score(y_test, y_pred)*100
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```

# Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
dtree_model = DecisionTreeClassifier(max_depth = 2).fit(x_train, y_train) 
```

```python
dp = dtree_model.predict(x_test) 
```

```python
cm = confusion_matrix(y_test, dp) 
cm
```

```python
from sklearn.metrics import classification_report  
print(classification_report(y_test,dp))
```

```python
d=metrics.accuracy_score(y_test,dp)*100
print("Decision tree classifier model accuracy(in %):", metrics.accuracy_score(y_test,dp)*100)
```

# SVM (Support vector machine) classifier

```python
from sklearn.svm import SVC 
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(x_train, y_train) 
```

```python
sp = svm_model_linear.predict(x_test) 
```

```python
s = svm_model_linear.score(x_test, y_test)*100
s
```

```python
svm_cm = confusion_matrix(y_test, sp)
svm_cm
```

```python
from sklearn.metrics import classification_report  
print(classification_report(y_test,sp))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors =7).fit(x_train, y_train) 
```

```python
k = knn.score(x_test, y_test)*100
k
```

```python
kp = knn.predict(x_test)  
```

```python
knn_cm = confusion_matrix(y_test, kp) 
knn_cm
```

```python
from sklearn.metrics import classification_report  
print(classification_report(y_test,kp))
```

# random forrest 


```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
```

```python
clf.fit(x_train,y_train)
```

```python
rd_y_pred=clf.predict(x_test)
```

```python
r=metrics.accuracy_score(y_test, rd_y_pred)*100
print("random forest Accuracy:",metrics.accuracy_score(y_test, rd_y_pred))
```

```python
from sklearn.metrics import classification_report  
print(classification_report(y_test,rd_y_pred))
```
